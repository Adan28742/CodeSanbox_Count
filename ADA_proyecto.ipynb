# Folium
Libreria para visualizar los datos

```python
import folium


listapx =  ((-12.0990933,-76.9692345 ),(-12.0992089, -76.9692794 ),(-12.0996268, -76.9694398 ))
coordinate = (-12.0990933,-76.9692345 )
m = folium.Map(
            tiles='Stamen Terrain',
            zoom_start=15,
            location=coordinate
        )
folium.PolyLine(listapx).add_to(m)
```

```python
import osmnx as ox
```

```python
#centro de lugar
center = (-12.104671062938884, -76.96161737561454)
road = ox.graph_from_point(center,dist=2000,network_type='drive')
ox.plot_graph(road)
```

```python
ox.save_graph_shapefile(road,"roadesan.shp")
```

```python
ox.save_graphml(road,"graoml.xml")
```

```python
road.number_of_nodes()
```

```python
road.number_of_edges()
```

```python
road.edges()
```

```python
camino = ox.distance.shortest_path(road,10970170008,392154413)
```
